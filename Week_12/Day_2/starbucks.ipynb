# Question 5

Summarise the mean number of calories in each beverage category.

# Question 6

Harder - Which Beverage_prep type contains more than the average calories of all drinks? (Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
# Calculate average calories for all drinks
cal_average = drinks.Calories.mean()
cal_average
```

```python
# Calculate means for Beverage_prep
grouped_prep = drinks.groupby("Beverage_prep").mean().round()
grouped_prep
```

```python
# Filter to find those which have more than average
grouped_prep["Calories"] >= cal_average
```

Group your espresso_drinks data frame by the type of Beverage prep, and then find out the mean calories for each drink group.

# Question 10
Get the same grouped mean Calories values as above, but this time sorted in descending order. (Hint: so you're able to sort the resulting groupby by the 'Calories' column you will want to ensure you get back a dataframe and not a series)

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Define plot parameters
# Figure Size
plt.rcParams['figure.figsize'] = [5, 3]
# Dots per inch
plt.rcParams['figure.dpi'] = 120
# Font size
plt.rcParams['font.size'] = 8
```

```python
sns.scatterplot( x = "Calories", y = "Caffeine (mg)", data = drinks_content)
plt.title("Comparing caffeine to calories");
```

```python
sns.scatterplot( x = "Calories", y = " Sugars (g)", data = drinks_content)
sns.set_theme(style="whitegrid")
plt.title("Comparing sugar content to calories");
# No surprise there - sugar and calories are linked
```

```python
sns.scatterplot( x = "Calories", y = " Sugars (g)", hue = "Beverage_category",  data = drinks_content)
sns.set_theme(style="whitegrid")
plt.title("Comparing sugar content to calories")
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0, fontsize='8', title_fontsize='8');
```
