## Question 1
Read in all 4 credit card transaction datasets and clean up the column names.

CardBase.csv
CustomerBase.csv
FraudBase.csv
TransactionBase.csv

```python
import pandas as pd

card = pd.read_csv('data/CardBase.csv')
customer = pd.read_csv('data/CustomerBase.csv')
fraud = pd.read_csv('data/FraudBase.csv')
transaction = pd.read_csv('data/TransactionBase.csv')
```

```python
card.columns = card.columns.str.strip().str.lower()\
    .str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
card.head()
```

```python
customer.columns = customer.columns.str.strip().str.lower()\
.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
customer.head()
```

```python
fraud.columns = fraud.columns.str.strip().str.lower()\
    .str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
fraud.head()
```

```python
transaction.columns = transaction.columns.str.strip().str.lower()\
    .str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
transaction.head()
```

## Question 3

Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
fraud_details = fraud.merge(transaction, how = "outer", on = "transaction_id")
```

```python
fraud_details
```

There are 10000 records in the joined table, the same as in the transaction table which tells us that all of the fraud transactions are contained in the transaction database.

## Question 1

Can you join all 4 credit card datasets together so that you’re left with a dataset that looks like below with 109 rows and 12 columns?

```python
# Merge fraud and transactions
step_one = fraud.merge(transaction, how = "left", on = "transaction_id")
step_one
```

```python
# Add in card
step_two = step_one.merge(card, how = "left", left_on = "credit_card_id",
                           right_on = "card_number") 
step_two
```

```python
# Add in customer
step_three = step_two.merge(customer, how = "left", on = "cust_id")
step_three
```

```python
# Reorder columns
final_dataset = step_three[["card_number", "card_family", "credit_limit",
                       "cust_id", "age", "customer_segment",
                       "customer_vintage_group", "transaction_id",
                       "transaction_date", "transaction_value", 
                       "transaction_segment", "fraud_flag", "credit_card_id"]]
del final_dataset['credit_card_id']

final_dataset
```
